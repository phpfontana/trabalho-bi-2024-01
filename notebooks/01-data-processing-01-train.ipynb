```python
PATH = '../data/processed/train_FD001.csv'

# Loading the processed dataset
data_processed = pd.read_csv(PATH)
```

```python
def remaining_useful_life(data, column):
    rul_data = data.copy()

    time_variable_max = rul_data[column].max()

    rul_data['RUL'] = time_variable_max - rul_data[column]

    return rul_data['RUL']

    
def make_multiclass_classification(data, column, thresholds, labels, categorical=False):
    data_multiclass = data.copy()

    bins = [-float('inf')] + thresholds + [float('inf')]
    
    if categorical:
        labels = list(labels.values())
    else:
        labels = list(labels.keys())

    data_multiclass[column] = pd.cut(data_multiclass[column], bins=bins, labels=labels)

    return data_multiclass[column]
```

```python
units_ = data_processed['unit_number'].unique()

print(f"Number of units: {len(units_)}")
```

```python
subsets = []

thresholds = [50, 125, 200]
labels = {0: 'urgent', 1: 'short', 2: 'medium', 3: 'long'}

for unit in units_:
    # Selecting the subset of the data by unit
    subset = data_processed[data_processed['unit_number'] == unit]

    subset['RUL'] = remaining_useful_life(subset, 'time_in_cycles')

    subset['maintanance_urgency'] = make_multiclass_classification(subset, 'RUL', thresholds, labels)

    # Appending the subset to the subsets list
    subsets.append(subset)
    

# Concatenating the subsets
data_prod = pd.concat(subsets)

data_prod.dropna(inplace=True)

data_prod.reset_index(drop=True, inplace=True)

data_prod.head(10)
```

```python
print("Dataset shape before adding rolling statistics")
print(f"rows: {data_processed.shape[0]}")
print(f"columns: {data_processed.shape[1]}")
print("Dataset shape after adding rolling statistics")
print(f"rows: {data_prod.shape[0]}")
print(f"columns: {data_prod.shape[1]}")
```

```python
# Save production ready data
data_prod.to_csv('../data/production/train/train_FD001.csv', index=False)
```
